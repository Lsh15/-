```python
from urllib.request import urlopen
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.chrome.options import Options
import time
```

```python
###웹 드라이버 실행###
###크롬드라이버 주소 변경, 크롤링할 구글 맵스 가게 주소로 변경###

driver = webdriver.Chrome('../chromedriver_win32/chromedriver')
driver.get('https://search.naver.com/search.naver?sm=top_hty&fbm=0&ie=utf8&query=%EC%B5%9C%EC%8B%A0%EC%A1%B1%EB%B0%9C')
```

```python
###리뷰 모두 보기 클릭###

xpath = """//*[@id="main_pack"]/div[3]/div[2]/a"""
driver.find_element_by_xpath(xpath).click()
```

```python
###블로그 들어가기 클릭### - path 자리 찾기

xpath = """//*[@id="main_pack"]/div[3]/ul/li[1]/div[1]/a/span[2]"""
driver.find_element_by_xpath(xpath).click()
```

```python
# 네이버 블로그 내용 가져오기 -> 작성 내용이 안불러짐

html  = driver.page_source
soup = BeautifulSoup(html,'html.parser')
soup
```
