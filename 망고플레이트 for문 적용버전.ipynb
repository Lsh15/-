```python
from urllib.request import urlopen
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.chrome.options import Options
import time
import csv
import pandas as pd
###웹 드라이버 실행###
###크롬드라이버 주소 변경###

driver = webdriver.Chrome('chromedriver_win32/chromedriver')
driver.get('https://www.mangoplate.com/')

#############팝업광고때문에 여기까지마 나눠서 먼저 실행############
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

data = pd.read_csv("data123.csv",encoding = 'CP949')
data_name = data['가게 명칭']

review_all = []
    
for n in data_name:

    time.sleep(0.5)
    driver.get('https://www.mangoplate.com/search/' + n) 
    
    
    #검색 결과 중1번째꺼 클릭
    xpath2 = """/html/body/main/article/div[2]/div/div/section/div[3]/ul/li[1]/div/figure/a/div/img"""
    try:
        driver.find_element_by_xpath(xpath2).click()
    except:
        None

    #리뷰 더보기 8번 누르기
    try:
        xpath3 = """/html/body/main/article/div[1]/div[1]/div/section[3]/div[2]"""
        for i in range(8):
            time.sleep(2)   
            driver.find_element_by_css_selector(xpath3).click() 
    except:
        None
    
    
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    review = []
    assess = []

    raw_all = soup.find_all(class_="RestaurantReviewItem RestaurantReviewList__ReviewItem")
    
    for item in raw_all:
        assess.append(item.find(class_="RestaurantReviewItem__RatingText").get_text())

        review.append(item.find(class_="RestaurantReviewItem__ReviewText").get_text())


    review_all.append(n)
    for item2 in range(len(review)):
        review_all.append(review[item2].replace("\n","").strip())
```
